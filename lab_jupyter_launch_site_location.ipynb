```python
# Task 2: Mark the success/failed launches for each site on the map

# Create the base map centered on the average launch coordinates
launch_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Create marker cluster to group close markers
marker_cluster = MarkerCluster().add_to(launch_map)

# Add colored markers for each launch
for index, row in spacex_df.iterrows():
    location = [row['Lat'], row['Long']]
    outcome = 'Success' if row['class'] == 1 else 'Failure'
    marker_color = 'green' if row['class'] == 1 else 'red'

    folium.Marker(
        location=location,
        icon=folium.Icon(color=marker_color),
        popup=f"{row['Launch Site']} - {outcome}"
    ).add_to(marker_cluster)

# Display map
launch_map
   
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
import folium
from math import radians, cos, sin, asin, sqrt
from folium.features import DivIcon

# Haversine function to calculate distances
def calculate_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1 
    dlon = lon2 - lon1 
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Earth radius in km
    return c * r

# Launch site: CCAFS LC-40
launch_site = [28.562302, -80.577356]

# Define closest proximity points (sample coordinates; adjust as needed)
closest_points = {
    "City": [28.3922, -80.6077],
    "Railway": [28.5721, -80.5853],
    "Highway": [28.5634, -80.5708]
}

# Create map
proximity_map = folium.Map(location=launch_site, zoom_start=12)

# Add launch site marker
folium.Marker(
    location=launch_site,
    popup="Launch Site",
    icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
).add_to(proximity_map)

# Loop through each proximity type
for label, coord in closest_points.items():
    # Calculate distance
    dist = calculate_distance(launch_site[0], launch_site[1], coord[0], coord[1])

    # Add marker for proximity point
    folium.Marker(
        location=coord,
        popup=f"{label} ({dist:.2f} km)",
        icon=folium.Icon(color='gray', icon='info-sign')
    ).add_to(proximity_map)

    # Draw line from launch site to proximity
    folium.PolyLine(locations=[launch_site, coord], color='red', dash_array='5').add_to(proximity_map)

    # Add label with distance at midpoint
    mid_lat = (launch_site[0] + coord[0]) / 2
    mid_lon = (launch_site[1] + coord[1]) / 2
    folium.Marker(
        location=[mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{dist:.2f} KM</b></div>',
        )
    ).add_to(proximity_map)

# Display the map
proximity_map
```
